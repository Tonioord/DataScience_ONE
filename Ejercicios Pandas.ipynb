```python
url3 = 'https://gist.githubusercontent.com/ahcamachod/807a2c1cf6c19108b2b701ea1791ab45/raw/fb84f8b2d8917a89de26679eccdbc8f9c1d2e933/alumnos.csv'
est=pd.read_csv(url3)
est.head()
```

```python
est['Puntos_extras'] = est['Nota'] * 0.4
est['Notas_finales'] = est['Nota'] + est['Puntos_extras']
est.head()
```

```python
est['aprobado_final'] = est['Notas_finales'].apply(lambda x: True if x >= 7.0 else False)
est.head()
```

```python
seleccioncambios = (est['Aprobado'] == 'False') & (est['aprobado_final'] == True)
est[seleccioncambios]
```
